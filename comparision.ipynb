```python
import shutil
import numpy as np
import matplotlib.pyplot as plt
params = {
    "figure.figsize": [9, 6],
    "axes.labelsize": 14,
    "axes.titlesize": 18,
    "axes.titlepad": 15,
    "font.size": 16,
    "legend.fontsize": 12,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
    "text.usetex": True if shutil.which("latex") else False,
    "font.family": "serif",
    "xtick.minor.visible": True,
    "ytick.minor.visible": True,
    "xtick.top": True,
    "ytick.left": True,
    "ytick.right": True,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.minor.size": 2.5,
    "xtick.major.size": 5,
    "ytick.minor.size": 2.5,
    "ytick.major.size": 5,
    "axes.axisbelow": True,
    "figure.dpi": 200,
}
plt.rcParams.update(params)
```

```python
ringdown = np.load("../my-first-run/ringdown/searched_areas.npy")
multimodal_32 = np.load("../my-first-run/multimodal_32/searched_areas.npy")
multimodal_rfft_32 = np.load("../my-first-run/multimodal_rfft_32/searched_areas.npy")

fig, ax = plt.subplots(1,1)
ax.step(ringdown, np.arange(1, len(ringdown) + 1) / len(ringdown), where="post", label="Ringdown")
ax.step(multimodal_32, np.arange(1, len(multimodal_32) + 1) / len(multimodal_32), where="post", label="Multimodal")
ax.step(multimodal_rfft_32, np.arange(1, len(multimodal_rfft_32) + 1) / len(multimodal_rfft_32), where="post", label="Multimodal RFFT")
ax.set_xscale("log")
ax.set_xlabel("Searched Area (deg$^2$)")
ax.set_ylabel("Cumulative Probability")
ax.set_title("Searched Area Cumulative Distribution Function")
ax.legend()
plt.tight_layout()
plt.show()
```
